# Data Visualization

we can use many tool for visualization 
Here we are using MATPLOTLIB

```python
import matplotlib.pyplot as plt
```

```python
x=[3,8]    #it will be counted as x1   x2

x1 = [2,7]

y=[7,14]    #y1   y2

y1 = [8,15]
```

```python
#this is for line format representation

plt.xlabel("Time")
plt.ylabel("Distance")
plt.plot(x,y,label="speed of car")     #to plot x,y coordinates
plt.plot(x1,y1,label="speed of bike")   #label add karne se it won't display for that u need to do plt.legend()

plt.grid(color='blue')   #to draw grids   can also give attr

plt.legend()

plt.show()
```

```python
#data repersentaion in form of bar chart

plt.xlabel("Time")
plt.ylabel("Distance")
plt.bar(x,y,label="speed of car")     
plt.bar(x1,y1,label="speed of bike")  

plt.grid(color='blue')   #to draw grids   can also give attr

plt.legend()

plt.show()


#if u want to plot line u can also do that by just using plot()
```
